Подготовка основных датасетов для анализа

```python
import pandas as pd
import numpy as np
import random
```

```python
# Основные таблицы генерируются с фиктивными данными на базе объекта Python - Faker Class.
import faker
```

```python
# Создадим экземпляр класса Faker
fake = faker.Faker()
```

```python
data = [
 { "Name": fake.name(), "Birthday":fake.date_of_birth(None, 18, 85),
  "Sex" : fake.word(ext_word_list=['M', 'F', 'N']),
  'email': fake.word() + '@gmail.com',
  "Address":fake.address(), "Company": fake.company(),
  "Profession":fake.job(), "Salary_USD$": np.random.randint(1500,5000),
  "foundation_date":fake.date_of_birth(None, 3, 5)
 }
 for i in range(8059)
 ]
```

```python
founders = pd.DataFrame(data = data)
founders
```

```python
world_cities = pd.read_csv('cities (1).csv')
world_cities.sample(5)
```

```python
# создаем фрейм со списком только по стране США (US)
US_cities = world_cities[world_cities.country_code == 'US']
US_cities
```

```python
# удаляем лишние колонки
US_cities = US_cities.drop(columns=['country_id', 'country_code', 'country_name', 'latitude', 'longitude', 'wikiDataId'])
US_cities
```

```python
# переименуем колонки
US_cities = US_cities.rename(columns={'id':'city_id', 'name':'city_name'})
US_cities
```

```python
# создадим рандомный и повторяющийся список городов США (по штатам)
US_cities = US_cities.sample(n=16390, replace = True).reset_index()
US_cities
```

```python
US_cities=US_cities.drop(columns=['index'])
US_cities
```

```python
# соединяем фрейм founders с таблицей US_states
founders = founders.join(US_cities, lsuffix='_x', rsuffix='_y')
founders
```

Наборы следующих открытых данных взяты из социальной сети Meetup, где пользователи объединяются в группы по интересам, например туризм, литература и настольные игры. Организаторы групп планируют удаленные или очные мероприятия, которые посещают члены группы. Предметная область Meetup имеет несколько моделей данных, включая группы, категории и города.
Файлы содержат образцы зарегистрированных групп Meetup. Каждая группа включает числовой идентификатор, название Funs Club, идентификатор укрупненной категории сферы интересов и идентификатор города.

```python
groups1=pd.read_csv('groups1.csv')
groups1
```

```python
groups2=pd.read_csv('groups2.csv')
groups2
```

Каждая группа имеет внешний ключ category_id. Информация о категориях хранится в файле category.csv. А уже в этом файле каждая строка хранит числовой идентификатор и название категории:

```python
categories = pd.read_csv("categories.csv")
categories
```

```python
# конкатенация фреймов groups1 и groups2
hobby_list = pd.concat(objs=[groups1, groups2], ignore_index = True)
hobby_list
```

```python
# выясняется, что в списке category_id фрейма hobby_list 33 уникальных значения
# а фрейм categories содержит таких только 26
hobby_list.category_id.nunique()
```

В дальнейшем можем получить проблему из-за наличия 7 category_id без соответствия названий category_name
Следует избавится от излишних category_id

```python
# соединим фрейм hobby с categories методом inner - только по совпадающим в обоих фреймах category_id
df = hobby_list.merge(categories, how = 'inner', on = 'category_id')
df
```

```python
# теперь списки category_id совпадают
df.category_id.nunique()
```

```python
# обновим фрейм hobby_list, удалив одновременно лишние колонки city_id и category_name
hobby_list = df.drop(columns=['city_id', 'category_name'])
hobby_list
```

```python
# количество уникальных hobbies
hobby_list.name.nunique()
```

```python
# поскольку в name есть дубли, удалим их для создания списка уникальных значений
hobby_list = hobby_list.drop_duplicates(subset = ["name"], keep = False)
hobby_list
```

```python
fee = [np.random.randint(0, 25) for i in range(8037)] 
```

```python
hobby_list['monthly_fee_USD$'] = pd.Series(data = fee)
hobby_list
```

```python
founders['Funs_club'] = fake.word(ext_word_list = [hobby_list.name])
founders
```

```python
# выясняем, что имеются дубли в именах владельце клубов - здесь оставим без изменений (представим. что один владелец
# организовал несколько клубов).
founders[['Funs_club','Name']].nunique()
```

```python
data_1 = [
 { "Name": fake.name(), "Birthday":fake.date_of_birth(None, 18, 85),
  "Sex" : fake.word(ext_word_list=['M', 'F', 'N']),
  'email': fake.word() + '@gmail.com',
  "Address":fake.address(), "Company": fake.company(),
  "Profession":fake.job(), "Salary_USD$": np.random.randint(1500,5000),
  "Entry_date":fake.date_of_birth(None, 3, 5),
  "group_id":np.random.choice(hobby_list.group_id)    
 }
 for i in range(460000)
 ]
```

```python
members = pd.DataFrame (data = data_1)
members
```

```python
# сразу уберем дубли в списке имен members - создадим таблицу со списком уникальных пользователей
members = members.drop_duplicates(subset = ['Name'], keep = False)
members
```

```python
# создадим рандомный список городов и штатов из списка US_cities 
US_cities1 = US_cities.sample(460000, replace = True).reset_index()
US_cities1
```

```python
US_cities1 = US_cities1.drop(columns=['index'])
US_cities1
```

```python
# присоединим получившийся фрейм к members
members = members.join(US_cities1, lsuffix='_x', rsuffix='_y')
members
```
