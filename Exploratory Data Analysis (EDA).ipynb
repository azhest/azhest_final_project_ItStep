```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
# Произведем загрузку датасетов и анализ типов содержащихся данных
founders = pd.read_csv('founders.csv')
```

```python
#создание функции выводящей все характеристики таблицы на экран
def first_check(dataset):

    print()
    print('Первые 3 строки таблицы')
    display(dataset.head(3))

    print()
    print('Последние 3 строки таблицы')
    display(dataset.tail(3))

    print()
    print('Информация о таблице')   
    print(dataset.info())

    print()
    print('Характеристики чисел в таблице')  
    display(dataset.describe())
      
    print()
    print('Переименование колонок')
def rename(dataset):
    dataset = dataset.rename(columns=lambda c: c.lower())
    return display(dataset.head(3))    
```

```python
first_check(founders)
rename(founders)
```

```python
# переименуем название колонок в нижнем регистре
founders = founders.rename(columns=lambda c: c.lower())
```

следует сделать следующие преобразования в объекте founders:
- преобразовать строки Birthday и foundation_date в метки даты
- преобразовать числа  city_id  и  state_id  в строки
- преобразовать строки Sex в категории
- столбец Funs_club содержит значения NaN

```python
founders['birthday'] = pd.to_datetime(founders['birthday'])
founders['foundation_date'] = pd.to_datetime(founders['foundation_date'])
founders['city_id'] = founders['city_id'].astype('str')
founders['state_id'] = founders['state_id'].astype('str')
founders['sex'] = founders['sex'].astype('category')
founders[:3]
```

```python
# В дальнейшем исследовании будут неинтересны строки с отсутствующими значениями funs_club, поэтому избавимся от них
founders = founders.dropna(subset = ["funs_club"])
founders
```
