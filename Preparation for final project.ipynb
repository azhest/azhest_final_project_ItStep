# Проект: "Анализ данных по основателям/владельцам "Клубов по интересам" (Funs Clubs) на территории США за последние 3 года"

```python
import pandas as pd
import numpy as np
import datetime as dt
import random
```

```python
# Основные таблицы генерируются с фиктивными данными на базе объекта Python - Faker Class.
import faker
```

```python
# Создадим экземпляр класса Faker
fake = faker.Faker()
```

```python
data = [
 { "Name": fake.name(), "Birthday":fake.date_of_birth(None, 18, 85),
  'email': fake.word() + '@gmail.com',
  "Address":fake.address(), "State_name":fake.state(), "Company": fake.company(),
  "Profession":fake.job(), "Salary_USD$": np.random.randint(1500,5000),
  "foundation_date":fake.date_of_birth(None, 3, 5)
 }
 for i in range(16390)
 ]
```

```python
founders = pd.DataFrame(data = data)
founders
```

```python
world_states = pd.read_csv('states.csv')
world_states
```

```python
# создаем фрейм со списком штатов только по стране США (US)
US_states = world_states[world_states.country_code == 'US']
US_states
```

```python
# переименуем колонку с названиями штатов
US_states = US_states.rename(columns = {'name':'State_name'})
US_states
```

```python
# соединяем фрейм founders с таблицей US_states
founders = founders.merge(US_states, how='inner', on='State_name')
founders
```

```python
# удаляем лишние колонки
founders = founders.drop(columns=['id', 'country_id', 'country_code', 'country_name', 'type', 'latitude', 'longitude'])
founders
```

Наборы следующих открытых данных взяты из социальной сети Meetup, где пользователи объединяются в группы по интересам, например туризм, литература и настольные игры. Организаторы групп планируют удаленные или очные мероприятия, которые посещают члены группы. Предметная область Meetup имеет несколько моделей данных, включая группы, категории и города.
Файлы содержат образцы зарегистрированных групп Meetup. Каждая группа включает числовой идентификатор, название Funs Club, идентификатор укрупненной категории сферы интересов и идентификатор города.

```python
groups1=pd.read_csv('groups1.csv')
groups1
```

```python
groups2=pd.read_csv('groups2.csv')
groups2
```

Каждая группа имеет внешний ключ category_id. Информация о категориях хранится в файле category.csv. А уже в этом файле каждая строка хранит числовой идентификатор и название категории:

```python
categories = pd.read_csv("categories.csv")
categories
```

Кроме того, каждая группа имеет внешний ключ city_id. Информация о городах хранится в наборе данных cities.csv. В нем каждый город имеет уникальный числовой идентификатор, название, штат и почтовый индекс

```python
cities=pd.read_csv('cities.csv', dtype={'zip':'string'})
cities
```

```python
# обращаем внимание, что данный фрейм ограничивает 
# дальнейшую аналитику данных списком только 6 штатов и 12 городов
cities[['state', 'city']].nunique()
```

```python
# конкатенация фреймов groups1 и groups2
hobby_list = pd.concat(objs=[groups1, groups2], ignore_index = True)
hobby_list
```

```python
# количество уникальных hobbies
hobby_list.name.nunique()
```

```python
# поскольку в name есть дубли, удалим их для создания списка уникальных значений
hobby_list = hobby_list.drop_duplicates(subset = ["name"])
hobby_list
```
